#### Ejercicio 18. 
Crea una nueva columna en el Data Frame que se llame Riesgo y tenga estos valores para cada grupo:

- 1 clase mujer = 1
- 2 clase mujer & 1 clase hombre = 2
- 3 clase mujer & 2 clase hombre = 3
- 3 clase hombre = 4

```python
titanic["Riesgo"] = 1
```

```python
titanic.loc[(titanic["Pclass"] == 2) & (titanic["Sex"] == "female"), "Riesgo"] = 2
titanic.loc[(titanic["Pclass"] == 3) & (titanic["Sex"] == "female"), "Riesgo"] = 3
titanic.loc[(titanic["Pclass"] == 1) & (titanic["Sex"] == "male"), "Riesgo"] = 2
titanic.loc[(titanic["Pclass"] == 2) & (titanic["Sex"] == "male"), "Riesgo"] = 3
titanic.loc[(titanic["Pclass"] == 3) & (titanic["Sex"] == "male"), "Riesgo"] = 4
```

```python
titanic.sample(5)
```

#### Ejercicio 19. 
Crea otra columna DesviacionFare con la diferencia de Fare y la media de todos los Fare del Data Frame.
Calcula el mínimo y el máximo de esta variable por cada combinación de sexo y clase 

```python
titanic["DesviacionFare"] = titanic["Fare"] - titanic["Fare"].mean()
```

```python
titanic.groupby(["Sex", "Pclass"])["DesviacionFare"].min()
```

```python
titanic.groupby(["Sex", "Pclass"])["DesviacionFare"].max()
```

```python
titanic.groupby(["Sex", "Pclass"])["DesviacionFare"].mean()
```
